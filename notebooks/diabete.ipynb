```python
import numpy as np
import pandas as pd
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
from sklearn import svm
from sklearn.metrics import accuracy_score
import pickle
```

```python
data = pd.read_csv("diabetes.csv")
```

```python
data.head()
```

```python
data.shape
```

```python
data.isnull().sum()
```

```python
data.describe()
```

```python
data["Outcome"].value_counts()
```

```python
X = data.drop(columns = "Outcome", axis = 1)
Y = data["Outcome"]
```

```python
print("Outcome: ", Y)
```

```python
scaler_diabete = StandardScaler()
```

```python
scaler_diabete.fit(X)
```

```python
standardized_data = scaler_diabete.transform(X)
```

```python
print(standardized_data )
```

```python
X = standardized_data
Y = data['Outcome']
```

```python
print(X)
print(Y)
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(X,Y, test_size = 0.2, stratify=Y, random_state=42)
```

```python
print(X.shape, X_train.shape, X_test.shape)
```

```python
classifier = svm.SVC(kernel='linear')
```

```python
classifier.fit(X_train, Y_train)
```

```python
X_train_prediction = classifier.predict(X_train)
training_data_accuracy = accuracy_score(X_train_prediction, Y_train)
```

```python
print('Accuracy score of the training data : ', training_data_accuracy)
```

```python
X_test_prediction = classifier.predict(X_test)
test_data_accuracy = accuracy_score(X_test_prediction, Y_test)
```

```python
print('Accuracy score of the test data : ', test_data_accuracy)
```

```python
input_data = (5,166,72,19,175,25.8,0.587,51)

# changing the input_data to numpy array
input_data_as_numpy_array = np.asarray(input_data)

# reshape the array as we are predicting for one instance
input_data_reshaped = input_data_as_numpy_array.reshape(1,-1)

# standardize the input data
std_data = scaler_diabete.transform(input_data_reshaped)
print(std_data)

prediction = classifier.predict(std_data)
print(prediction)

if (prediction[0] == 0):
  print('Do not worry, there is no diabetic')
else:
  print('Unfortunately, there is diabetic')
```

```python
with open('diabetes_model.pkl', 'wb') as model_file:
    pickle.dump(classifier, model_file)

with open('scaler_diabete.pkl', 'wb') as scaler_file:
    pickle.dump(scaler_diabete, scaler_file)

print("Model ve scaler başarıyla kaydedildi.")
```
